```python
#导入需要的包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Teaser绘制

```python
fs = 24
# 数据
sample_per_class_labeled = [1500, 899, 539, 323, 193, 116, 69, 41, 25, 15]
sample_per_class_unlabeled = [3000, 1798, 1078, 646, 387, 232, 139, 83, 50, 30]
# x = range(len(sample_per_class_labeled))
x = np.arange(0, len(acc_per_class_fixmatch)*2, 2)
width = 1
x1 = x - width/2
x2 = x + width/2

# 绘图
fig, ax = plt.subplots(figsize=(15,6)) # 新增 figsize 参数
ax.bar(x=x1, height=sample_per_class_unlabeled, width=0.92, color='#FFDFDF', alpha=0.6, label='Unlabeled', edgecolor='black', linewidth=0.3)
ax.bar(x=x2, height=sample_per_class_labeled, width=0.92, color='#FB929E', alpha=0.6, label='Labeled' , edgecolor='black', linewidth=0.3)
# 隐藏右侧和上方的框线
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.legend(fontsize=fs, bbox_to_anchor=(1, 1.1)) #显示图例
plt.xlabel('Class index', fontsize=fs)
plt.ylabel('Sample num', fontsize=fs)
plt.xticks(x, fontsize=fs) #设置x轴标签字体大小
plt.yticks(fontsize=fs) #设置y轴标签字体大小
# 设置x轴刻度
ax.set_xticks(x)
ax.set_xticklabels([str(i) for i in range(10)]) # 修改刻度标签内容

# 保存图像
plt.savefig('figures/teaser_long_tailed_distribution', dpi=300, bbox_inches='tight')
```

```python
fs = 24
# 数据
acc_per_class_fixmatch = [98.7, 99.5, 89.4, 83.1, 86.9, 60.6, 78.7, 58.1, 48.,  24.2] # r100
acc_per_class_ours = [92.3, 97.,  88.2, 76.8, 87.3, 71.,  85.7, 77.7, 87.6, 77.5] # r100
x = np.arange(0, len(acc_per_class_fixmatch)*2, 2)
width = 1
x1 = x - width/2
x2 = x + width/2

# 绘图
fig, ax = plt.subplots(figsize=(15,6)) # 新增 figsize 参数
ax.bar(x=x1, height=acc_per_class_fixmatch, width=0.92, color='#e8cefb', alpha=0.6, label='FixMatch', edgecolor='black', linewidth=0.3)
ax.bar(x=x2, height=acc_per_class_ours, width=0.92, color='#9689f3', alpha=0.6, label='Ours', edgecolor='black', linewidth=0.3)

# 隐藏右侧和上方的框线
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# 显示图例
ax.legend(fontsize=fs,loc='upper right', bbox_to_anchor=(1, 1.1)) # 修改位置为右上角

# 设置坐标轴标签字体大小
ax.set_xlabel('Class index', fontsize=fs)
ax.set_ylabel('Accuracy', fontsize=fs)
ax.tick_params(axis='both', which='major', labelsize=fs) # 修改刻度标签字体大小

# 设置x轴刻度
ax.set_xticks(x)
ax.set_xticklabels([str(i) for i in range(10)]) # 修改刻度标签内容

# 保存图像
plt.savefig('figures/per_class_acc.png', dpi=300, bbox_inches='tight')
```
